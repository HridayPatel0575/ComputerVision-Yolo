```python
cap =cv2.VideoCapture(0)
cap.set(3,1280)
cap.set(4,720)

model = YOLO('yolov8l.pt').to('cuda')

print(model.device)

while True:
    success,img = cap.read()
    result = model(img,stream=True)
    for r in result:
        boxes = r.boxes
        for box in boxes:
            x1,y1,x2,y2=box.xyxy[0]
            x1,y1,x2,y2 = int(x1),int(y1),int(x2),int(y2)
            w,h = x2-x1,y2-y1
            cvzone.cornerRect(img,(x1,y1,w,h))
            # cv2.rectangle(img,(x1,y1),(x2,y2),(255,0,255),3)
            conf = math.ceil((box.conf[0]*100))/100
            # cv2.putText(img,f'{conf}',(x1,y1-20),fontFace=1,fontScale=4,color=(255,0,255))
            cls=int(box.cls[0])
            cvzone.putTextRect(img,f'{classNames[cls]} {conf}',(max(0,x1),max(30,y1-20)))

    cv2.imshow('image',img)
    cv2.waitKey(1)
    
```
